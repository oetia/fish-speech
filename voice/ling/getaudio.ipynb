```python
import requests
# for i in range(1, 36):
for i in range(36, 45):
  url = f"https://raw.githubusercontent.com/Aceship/Arknight-voices/main/voice_cn/char_2023_ling/CN_0{str(i).zfill(2)}.mp3"
  file_name = url.split('/')[-1]
  print(f"Downloading {file_name}...")
  r = requests.get(url)
  if r.status_code == 200:
    with open("./aceship/" + file_name, 'wb') as f:
      f.write(r.content)
  else:
    print(f"Tch. Failed to download {file_name}. Status code: {r.status_code}")
```
